```python
LEVEL = '1000x10/'
PATH_FINAL_TEST_SET = './' + LEVEL + '/Final_Test_Set/'
PATH_FINAL_TEST_FEATURES = './' + LEVEL + '/Final_Test_Features/'
```

```python
def readlist(file_name) :
    with open(file_name, 'r') as f :
        line = f.readline()
        # print line
        len_list = line.split(' ') 
        len_list = map(int, len_list)
        # print len_list
    
    return len_list
```
